# ch02/and_gate.py

```python
import numpy as np


def and_gate(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1


if __name__ == '__main__':
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        y = and_gate(xs[0], xs[1])
        print(str(xs) + " -> " + str(y))
```

# ch02/nand_gate.py

```python
import numpy as np


def nand_gate(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1


if __name__ == '__main__':
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        y = nand_gate(xs[0], xs[1])
        print(str(xs) + " -> " + str(y))
```

# ch02/or_gate.py

```python
import numpy as np


def or_gate(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1


if __name__ == '__main__':
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        y = or_gate(xs[0], xs[1])
        print(str(xs) + " -> " + str(y))
```

# ch02/xor_gate.py

```python
# 依存チェック：このセルだけ先に実行すると NameError になるのを防ぐ
_required = ("and_gate", "nand_gate", "or_gate")
_missing = [name for name in _required if name not in globals()]
if _missing:
    raise RuntimeError(
        f"先に {_missing} の定義セルを実行してください（ノートブックは実行順に依存します）"
    )


def xor_gate(x1, x2):
    s1 = nand_gate(x1, x2)
    s2 = or_gate(x1, x2)
    out = and_gate(s1, s2)
    return out


if __name__ == '__main__':
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        y = xor_gate(xs[0], xs[1])
        print(str(xs) + " -> " + str(y))
```
